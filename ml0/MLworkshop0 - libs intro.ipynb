### Numpy

```python
import numpy as np
```

#####  Important attributes

```python
a = np.arange(20)
print(a)
print(type(a))
print(a.shape)
print(a.ndim)
print(a.dtype)
print(a.size)
print(a.itemsize)
```

```python
a = np.arange(20, dtype=np.float32)
print(a)
print(type(a))
print(a.shape)
print(a.ndim)
print(a.dtype)
print(a.size)
print(a.itemsize)
```

##### Array creations 

```python
a = np.array(0)
print(a, a.shape)
a = np.array([1,2,3])
print(a, a.shape, '\n')
a = np.array([[1,2,3], [1,2,3]])
print(a, a.shape, '\n')
a = np.array([[[1,2,3], [1,2,3]], [[1,2,3], [1,2,3]]])
print(a, a.shape)
```

```python
print(np.zeros((3,4))) # remember the double parentheses, you need to insert tuple into fun
print(np.ones((3,4)))
print(np.ones((3,4)) * 8)
print(np.empty((3,4)))
print(np.empty((3,5)))
```

```python
print(np.full((3,4), 7, dtype=np.int64))
print(np.eye(6))
print(np.random.random((4,5)))

a = np.empty((3,2)) 
print(np.zeros_like(a))
```

```python
a = np.arange(-5,15).reshape((4,5))
b = np.arange(1,21).reshape((4,5))
print(a)
print(b)
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)
```

```python
print(np.sin(a))
print(np.log10(b))
print(np.abs(a))
print(a < 10)
```

```python
print(a)
print(a.T)
print(np.dot(a, a.T))
print(np.dot(a.T, a))
print(a @ a.T) #python >=3.5 feature
```

```python
print(a)
print(a.sum(axis=0))
print(a.min(axis=0))
print(a.min(axis=1))
```

##### Slicing 

```python
a = np.arange(20).reshape((4,-1))
print(a)
print(a[2])
print(a[2:])
print(a[:,:])
print(a[1,:])
```

```python
a = np.arange(4).reshape((2,-1))
print(np.vstack((a,a)))
print(np.hstack((a,a)))
print(np.concatenate((a[0],a[0])))
```

```python
a = np.arange(20)
print(a)
b = a % 3 == 2
print(b)
print(a[b])
```
